Percebe-se que as classes da coluna-alvo estão desbalanceadas. Por isso, é necessário fazer um balanceamento da base para evitar que o modelo fique enviezado.

Também serão removidos registros com `koi_disposition` classificados como **CANDIDATE**, já que o modelo classificará apenas em **CONFIRMED** ou **FALSE POSITIVE**.

Os atributos `kepid`, `kepoi_name` e `kepler_name` possuem muitos valores únicos. Logo, eles podem ser removidos da base.

Além disso, os atributos `koi_pdisposition` e `koi_score` também podem ser desconsiderados, visto que `koi_pdisposition` trabalha apenas com possíveis classificações, e `koi_score` depende de `koi_pdisposition`.

Também serão removidos atributos de incerteza, ou seja, aquelas com rótulos que contenham **_err**.

Com exceção do balanceamento, essa limpeza foi feita no arquivo **cleaning_data.ipynb**

```python
import matplotlib.pyplot as plt
```

```python
# Verificação da presença de outliers entre as colunas numéricas, por boxplot
for column in df.select_dtypes("number").columns:

    fig, ax = plt.subplots()

    ax.boxplot(df[column])
    ax.set_title(column)

    plt.show()
```

Quase todas as colunas apresentam outliers, mas em grande quantidade. Isso significa que, provalmente, esses valores não são errôneos. Logo, não serão removidos da base

No entanto, a coluna `koi_fpflag_nt` apresenta um valor suspeito, muito maior que os demais. 

```python
df["koi_fpflag_nt"].value_counts()
```

```python
# Removendo o outlier com valor 465 da coluna "koi_fpflag_nt"

df = df[df["koi_fpflag_nt"] != 465]
```

```python
# Verificando os valores de "koi_fpflag_ec"

df["koi_fpflag_ec"].value_counts()
```

Apesar de o boxplot aparentar exibir um único valor da coluna "koi_fpflag_ec" como sendo 1, na realidade existem 1134 instâncias com valor 1 nessa coluna.

```python
# Verificando novamente a quantidade de valores únicos de cada coluna

df.nunique()
```

```python
# Classificando as colunas em categóricas (Não ordenadas), coluna alvo ou outras colunas

target_column = ["koi_disposition"]
unordered_categoric_columns = [column for column, value in df.nunique().items() if column != target_column[0] and value < 20]
other_columns = [column for column in df.columns if column not in (target_column + unordered_categoric_columns)]
```

```python
unordered_categoric_columns
```

```python
other_columns
```

```python
# Definindo as colunas categoricas como do tipo CategoricalDtype

for column in unordered_categoric_columns:
    df[column] = df[column].astype("category")
```

```python
df.info()
```

```python
# Analisando o comportamento das flags para cada classificação

for column in filter(lambda x: "flag" in x, df.columns):

    display(df.groupby([column, "koi_disposition"]).size())
```

Percebe-se que o valor 1 nas colunas de flag estão presentes, geralmente, em KOIs classificados como falso positivo. Assim, essas colunas podem contribuir para a predição do modelo.
